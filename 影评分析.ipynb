```python

from bs4 import BeautifulSoup
import requests
import pandas as pd

urls = ['http://movie.mtime.com/16608/reviews/short/new-{}.html'.format(str(i)) for i in range(1,11)]
 # print(urls)
def get_titles(urls,data = None):
    web_data = requests.get(urls)
    soup = BeautifulSoup(web_data.text, 'lxml')
    titles = soup.select(' h3 > a > em > strong')
    review=[]
    
    for p in soup.find_all('h3'):
        review.append(p.text)
        
    rating= [] # ratings list, there are empty ratings, so we can just just find all h3, we need to find all p
    for p in soup.find_all('p',{'class':"mt6 px12 clearfix"}):
        rating.append(p.text)
        
    # now get all the ratings
    rating = [x.split(' ')[-1] for x in rating]
    return (rating,review)

rating_all = []
review_all = []

for titles in urls:
    rat_rev = get_titles(titles)
    rating_all = rating_all + rat_rev[0]
    review_all = review_all + rat_rev[1]
    
```

```python
df = pd.DataFrame({"ratings":rating_all,"reviews": review_all})
df

# df.to_csv('新警察故事.csv')
```

```python
#clustering
from sklearn.cluster import KMeans
# X is a table with 2 volumns(features)
feature_list=['ratings','reviews']
X=df[].dropna().values
kmeans=KMeans(n_clusters=10, random_state=0).fit(X)
print (X)
print(set(kmeans.labels_))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import jieba
import pyLDAvis
import pyLDAvis.sklearn
```

```python


# 读取评论数据
hComments = []
with open('2046.txt', 'r', encoding="utf-8") as f1:
    for line in f1:
        hComments.append(" ".join(jieba.cut(line)))


df = pd.DataFrame(hComments)

# 关键词提取和向量转化
tfVectorizer = CountVectorizer(strip_accents = 'unicode',
                                max_features = 1000,
                                max_df = 0.5,
                                min_df = 10
                                )
tf = tfVectorizer.fit_transform(df[0])

# 初始化lda
lda = LatentDirichletAllocation(n_topics = 3,
                                max_iter =50,
                                learning_method = 'online',
                                learning_offset = 50,
                                random_state = 0)
lda.fit(tf)  # 训练

# 可视化lda
data = pyLDAvis.sklearn.prepare(lda, tf, tfVectorizer)
pyLDAvis.show(data)
```
